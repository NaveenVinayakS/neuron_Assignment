```python
Q1. What is the purpose of Python's OOP?

A1. Python's Object-Oriented Programming (OOP) allows you to structure your code using objects and classes. The main purposes of OOP in Python are encapsulation, inheritance, and polymorphism. It helps in organizing code into reusable and maintainable components.

Q2. Where does an inheritance search look for an attribute?

A2. In Python, when you access an attribute of an object, it first searches in the instance itself. If not found, it looks in the class of the instance and then in its superclass(es) in the order specified by the method resolution order (MRO).

Q3. How do you distinguish between a class object and an instance object?

A3. Class objects are created when you define a class, and instance objects are created when you instantiate a class. Class objects represent the class itself, while instance objects represent individual instances of the class.

class MyClass:
    class_attribute = 10

obj1 = MyClass()  # obj1 is an instance object
obj2 = MyClass()  # obj2 is another instance object

Q4. What makes the first argument in a classâ€™s method function special?

A4. The first argument in a class's method function is typically named self. It represents the instance of the class on which the method is called. This argument allows you to access and manipulate instance-specific data and methods.

class MyClass:
    def my_method(self):
        print("This is a method of MyClass")

obj = MyClass()
obj.my_method()  # 'self' refers to 'obj'


Q5. What is the purpose of the init method?

A5. The __init__ method is a special method in Python that is automatically called when an instance of a class is created. It is used for initializing instance-specific attributes and performing setup tasks when an object is created.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Alice", 30)


Q6. What is the process for creating a class instance?

A6. To create an instance of a class in Python, you call the class as if it were a function, passing any required arguments to the __init__ method (if defined).

class MyClass:
    def __init__(self, value):
        self.value = value

obj = MyClass(42)  # Creating an instance of MyClass with value 42


Q7. What is the process for creating a class?

A7. To create a class in Python, you use the class keyword, followed by the class name and a colon. You define class attributes and methods within the class block.

class MyClass:
    class_attribute = 10

    def my_method(self):
        print("This is a method of MyClass")
Q8. How would you define the superclasses of a class?

A8. Superclasses (also known as parent classes or base classes) of a class are defined by specifying them in parentheses after the class name when defining the class. Multiple superclasses can be listed, and the class inherits attributes and methods from all its superclasses.

class ParentClass:
    def parent_method(self):
        print("This is a method of ParentClass")

class ChildClass(ParentClass):
    def child_method(self):
        print("This is a method of ChildClass")

obj = ChildClass()
obj.parent_method()  # Accessing a method from the superclass
```
