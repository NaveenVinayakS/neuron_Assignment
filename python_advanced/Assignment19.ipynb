```python
#Q1. Define the relationship between a class and its instances. Is it a one-to-one or a one-to-many partnership, for example?

Answer: The relationship between a class and its instances is a one-to-many partnership. A class serves as a blueprint or
    template for creating multiple instances, and each instance can have its own unique state and behavior while adhering 
    to the class's structure.

#Q2. What kind of data is held only in an instance?

Answer: Data held only in an instance is typically instance-specific data, which includes attributes or variables that are 
    unique to each instance of the class. These attributes store the state or characteristics of a particular instance.

#Q3. What kind of knowledge is stored in a class?

Answer: Knowledge stored in a class includes class-specific data and methods. Class variables store data shared among all 
    instances of the class, and class methods define behavior or operations that are common to all instances of the class.

#Q4. What exactly is a method, and how is it different from a regular function?

Answer: A method is a function that is associated with a class. The key difference between a method and a regular function
    is that a method is defined within a class and operates on class instances. Methods can access and modify 
    instance-specific data, whereas regular functions are standalone and do not have access to instance data unless it is
    passed as an argument.

#Q5. Is inheritance supported in Python, and if so, what is the syntax?

Answer: Yes, inheritance is supported in Python. Inheritance allows a class to inherit attributes and methods from another class. 
class ChildClass(ParentClass):
    # ChildClass inherits from ParentClass
    # Define additional attributes and methods specific to ChildClass

#Q6. How much encapsulation (making instance or class variables private) does Python support?

Answer: Python supports encapsulation to a certain extent through naming conventions. Variables or methods can be made 
    "private" by prefixing their names with a single underscore (e.g., _my_variable). However, this is a convention, and 
    it does not provide true access control. Its a signal to other developers that a variable or method is intended for
    internal use, but it can still be accessed from outside the class.

#Q7. How do you distinguish between a class variable and an instance variable?

Answer: Class variables are shared among all instances of a class and are defined within the class but outside any methods.
    Instance variables are specific to each instance and are defined within methods or the class __init__ constructor. 
    You can access class variables using the class name (e.g., ClassName.class_variable) and instance variables using an
    instance of the class (e.g., instance.instance_variable).

#Q8. When, if ever, can self be included in a class's method definitions?

Answer: self should be included as the first parameter in a class method definitions. It represents the instance that the
    method is operating on and allows you to access instance-specific data and methods within the method.

#Q9. What is the difference between the __add__ and the __radd__ methods?

Answer: The __add__ method is used to define the behavior of the addition operator (+) when the object is on the left side
    of the operator. The __radd__ method is used to define the behavior of the addition operator when the object is on the
    right side of the operator. These methods allow you to customize how objects of your class interact with the addition 
    operator.

#Q10. When is it necessary to use a reflection method? When do you not need it, even though you support the operation in question?

Answer: A reflection method, such as __str__ or __repr__, is necessary when you want to provide a custom string 
    representation of an object. You dont need it if the default string representation is sufficient for your use case.

#Q11. What is the __iadd__ method called?

Answer: The __iadd__ method is called when the in-place addition operator (+=) is used on an object. It allows you to
    define the behavior of the += operator for instances of your class.

#Q12. Is the __init__ method inherited by subclasses? What do you do if you need to customize its behavior within a subclass?

Answer: The __init__ method is not automatically inherited by subclasses, but a subclass can choose to override it.
    To customize the behavior of __init__ within a subclass, you can define a new __init__ method in the subclass, and
    it should call the superclass's __init__ method if you want to retain the superclass's behavior while adding or 
    modifying functionality in the subclass's constructor.
```
