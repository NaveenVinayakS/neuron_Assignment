```python
Q1. Text files and binary files differ in how they store data. Text files contain human-readable characters encoded in a 
specific format, like ASCII or UTF-8, with line breaks and text formatting. Binary files, on the other hand, store data in
a non-human-readable format, preserving the exact sequence of bytes. Text files are suitable for storing and sharing 
information, while binary files are used for non-text data like images, executables, and more, where the exact binary structure
must be maintained.

Q2. Text files are better for storing and sharing human-readable textual data, like configuration files, source code, or
logs. Use binary files when you need to preserve the exact binary structure, such as for multimedia files or proprietary 
data formats.

Q3. When using binary operations to read and write a Python integer directly to disk, you may encounter issues related to 
endianness (byte order) and data size. Different platforms have different byte orders, which can lead to compatibility 
problems when reading data across systems. Additionally, the size of an integer can vary, causing issues when reading or
writing integers in binary without proper consideration of size and format.

Q4. Using the 'with' keyword in Python to open files (e.g., with open('file.txt', 'r') as file) is beneficial because it
automatically handles file closing. It ensures that the file is properly closed when youre done with it or if an exception
occurs, making code cleaner and more robust.

Q5. When reading a line of text in Python, it includes the trailing newline character. When writing a line of text, Python
does not append a newline character by default. You need to explicitly add '\n' if you want a newline.

Q6. Random access operations are enabled by file operations like 'seek()' and 'tell()' in Python. 'seek()' allows you to
move the file pointer to a specific position, while 'tell()' tells you the current position in the file, enabling you to 
read or write data at arbitrary locations within the file.

Q7. Yoll use the 'struct' package most when you need to work with binary data and handle various data types, such as 
integers, floats, and complex data structures. Its commonly used in low-level programming, networking, and working with 
binary file formats.

Q8. Pickling is the best option when you need to serialize and deserialize Python objects to and from a binary format for 
data storage, transfer, or inter-process communication.

Q9. You should use the 'shelve' package when you need to store Python objects in a persistent dictionary-like structure,
allowing you to save and retrieve complex data structures with keys. It useful for managing data that you want to persist
between runs of your Python programs.

Q10. A special restriction when using the 'shelve' package is that the keys in the shelve database must be strings. 
Unlike other data dictionaries, which allow a wider range of data types as keys, 'shelve' restricts keys to strings for
serialization purposes.
```
