```python
#Q1. What is the distinction between a numpy array and a pandas data frame? Is there a way to convert between the two if there is?

Answer: Numpy arrays and Pandas DataFrames are both data structures commonly used in Python for data manipulation, but they
    have some key differences.

Numpy Array:
Numpy arrays are homogeneous, meaning they store elements of the same data type.
They are mainly used for numerical and mathematical operations.
Numpy arrays are more memory-efficient and faster for numerical computations.
Pandas DataFrame:
Pandas DataFrames are heterogeneous, allowing for different data types in each column.
They are used for working with structured data and are ideal for data analysis and manipulation.
Pandas DataFrames provide powerful data indexing and manipulation features.
To convert between the two, you can use the following methods:

To convert a Pandas DataFrame to a Numpy array: df.to_numpy() or df.values.
To convert a Numpy array to a Pandas DataFrame: pd.DataFrame(array), where pd is the Pandas library.

#Q2. What can go wrong when a user enters a stock-ticker symbol, and how do you handle it?

Answer: Several issues can arise when a user enters a stock-ticker symbol:

Incorrect Ticker Symbol: Users might enter a ticker symbol that doesnt exist or is misspelled. To handle this, you can
    validate the symbol against a reliable data source or API.

Missing Data: Some ticker symbols might not have available data for the specified time frame. In this case, you should 
    provide appropriate error handling and user feedback.

Data Format Issues: Users may input data in an incorrect format or with extra characters. Proper input validation and 
    formatting are essential to prevent issues.

API Errors: If youre using an API to fetch stock data, it may experience downtime or rate limits. Implement error handling
    and retries for such situations.

Handling these issues involves input validation, error handling, and providing clear feedback to the user when errors 
occur.

#Q3. Identify some of the plotting techniques that are used to produce a stock-market chart.

Answer: Stock-market charts can be created using various plotting techniques. Some common ones include:

Line Charts: Line charts are used to display the price or performance of a stock over time. They are straightforward and 
    can show trends.

Candlestick Charts: Candlestick charts display open, close, high, and low prices for a stock within a specific time period,
    providing a more detailed view of price movements.

OHLC Charts: OHLC (Open, High, Low, Close) charts are similar to candlestick charts but provide a simple bar representation
    for each time period.

Moving Averages: Plotting moving averages, such as the 50-day or 200-day moving average, can help identify trends and 
    support/resistance levels.

Volume Charts: Volume charts show the trading volume for a stock, helping to analyze the liquidity and interest in a
    particular stock.

Bollinger Bands: Bollinger Bands combine a moving average with volatility bands to visualize potential price volatility and
    reversals.

RSI (Relative Strength Index) Charts: RSI charts help identify overbought or oversold conditions, indicating potential
    trend reversals.

#Q4. Why is it essential to print a legend on a stock market chart?

Answer: Printing a legend on a stock market chart is essential for clarity and understanding. A legend provides labels for
    the various elements on the chart, helping viewers interpret the information. In the context of a stock market chart, 
    the legend typically explains what each line, color, or symbol represents. The importance of a legend includes:

Clear Interpretation: Without a legend, viewers may struggle to understand which line or data series corresponds to which 
    stock or data point.

Data Context: A legend provides context for the chart, ensuring that viewers can make informed decisions based on the 
    displayed data.

Avoiding Confusion: Stock charts often include multiple stocks, indices, or indicators. A legend prevents confusion by 
    clearly indicating which is which.

Professional Presentation: Including a legend enhances the professionalism and readability of the chart, making it more
    accessible to a wider audience.

#Q5. What is the best way to limit the length of a Pandas DataFrame to less than a year?

Answer: To limit the length of a Pandas DataFrame to less than a year, you can filter the data based on the date or time.
    Assuming your DataFrame has a date column
    
import pandas as pd
from datetime import datetime, timedelta

# Assuming 'date_column' is the name of your date column
# and 'df' is your DataFrame

# Set the end date as today
end_date = datetime.today()

# Calculate the start date by subtracting one year from the end date
start_date = end_date - timedelta(days=365)

# Filter the DataFrame to include only data within the past year
filtered_df = df[(df['date_column'] >= start_date) & (df['date_column'] <= end_date)]
```
