```python
# 1. What do REGION PROPOSALS entail?
Region proposals are a key concept in object detection. They involve generating candidate bounding boxes in an image that
potentially contain objects of interest. These proposals help reduce the computational cost of object detection by focusing 
on regions likely to contain objects, rather than evaluating the entire image. Various algorithms, like Selective Search
and EdgeBoxes, can be used to generate these region proposals.

# 2. What do you mean by NON-MAXIMUM SUPPRESSION? (NMS)
Non-Maximum Suppression is a post-processing technique used in object detection and computer vision. It aims to eliminate 
redundant or overlapping bounding boxes generated by object detection algorithms. NMS identifies the bounding box with the
highest confidence score (e.g., probability of containing an object) and removes overlapping boxes with lower scores. This 
ensures that only the most confident and non-overlapping boxes are retained.

# 3. What exactly is mAP?
mAP stands for Mean Average Precision. It is a metric used to evaluate the performance of object detection and image 
classification models. mAP takes into account both precision and recall across multiple object categories. It measures the
quality of object localization and the ability to classify objects correctly. A higher mAP indicates better model 
performance.

# 4. What is frames per second (FPS)?
Frames per second (FPS) is a measure of a system's or device's ability to display or process a certain number of video 
frames in one second. In the context of computer vision and object detection, it indicates how many frames an algorithm or
system can process or display in a single second. Higher FPS values are generally desirable for real-time video analysis
and applications.

# 5. What is an IOU (INTERSECTION OVER UNION)?
Intersection over Union (IoU) is a metric used to evaluate the accuracy of object detection algorithms. It measures the 
overlap between the predicted bounding box and the ground truth bounding box for an object. IoU is calculated as the ratio 
of the area of the intersection between the two boxes to the area of their union. It is often used in non-maximum 
suppression to determine which bounding boxes should be retained.

# 6. Describe the PRECISION-RECALL CURVE (PR CURVE)
The Precision-Recall curve (PR curve) is a graphical representation of the trade-off between precision and recall for a 
binary classification model, such as object detection. Precision measures the accuracy of positive predictions, while 
recall measures the ability to capture all positive instances. The PR curve shows how these metrics change at various 
classification thresholds. A good model will have a PR curve that is closer to the top-right corner, indicating high 
precision and high recall.

# 7. What is the term "selective search"?
Selective Search is an object proposal generation method used in object detection. It is a region proposal algorithm that
identifies potential object locations by grouping pixels into segments based on color, texture, and other image 
characteristics. Selective Search generates a diverse set of candidate bounding boxes, which can be used as 
input to object detection models.

# 8. Describe the R-CNN model's four components.

The R-CNN (Region-based Convolutional Neural Network) model consists of four main components:
Region Proposal: Initial region proposals are generated using methods like Selective Search.
Feature Extraction: The selected regions are passed through a CNN to extract features.
Object Classification: These features are used to classify objects within the proposed regions.
Bounding Box Regression: The model refines the bounding box coordinates for precise object localization.

# 9. What exactly is the Localization Module?
The Localization Module in an object detection model is responsible for refining the bounding box coordinates for the 
proposed objects. It takes the extracted features from the region proposal and adjusts the bounding box to align with the
objects actual location within the proposed region. This module is critical for accurate object localization.

# 10. What are the R-CNN DISADVANTAGES?
R-CNN models have several disadvantages, including:
Slow Inference: R-CNNs are computationally expensive and slow for real-time applications due to their multi-stage processing.
High Memory Usage: They require significant memory to store feature maps for each region proposal, which can be inefficient.
Training Complexity: Training R-CNNs involves multiple stages and can be complex and time-consuming.
Lack of End-to-End Training: R-CNNs involve separate training stages for region proposal and object classification, which can 
make the model less unified and harder to optimize.
```
